```python
# ==========================================
# NAMED ENTITY RECOGNITION (NER) TAGS - NLTK
# (Used by nltk.ne_chunk)
# ==========================================

# PERSON        -> Names of people (e.g., "Shah Jahan", "John")
# ORGANIZATION  -> Companies, institutions (e.g., "Google", "UN")
# GPE           -> Geo-Political Entity (countries, cities, states)
# LOCATION      -> Non-political places (e.g., "Himalayas")
# FACILITY      -> Buildings, airports, roads (e.g., "Taj Mahal")
# GSP           -> Geopolitical groups (rarely used)

# DATE          -> Dates (e.g., "January 1, 2020")
# TIME          -> Time expressions (e.g., "10 AM")
# MONEY         -> Monetary values (e.g., "$100")
# PERCENT       -> Percentage (e.g., "50%")
```

```python
import nltk
nltk.download('punkt')
nltk.download('maxent_ne_chunker')
nltk.download('words')
nltk.download('averaged_perceptron_tagger')
```

```python
import nltk

sentence = "Taj Mahal is in India and was built by Shah Jahan in 1632"

words = nltk.word_tokenize(sentence)
pos_tags = nltk.pos_tag(words)
ner = nltk.ne_chunk(pos_tags)

print(ner)
```
