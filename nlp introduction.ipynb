
## 🔹 What NLP Actually Does

It helps machines work with text or speech just like humans do.
Examples:

* Chatbots (like ChatGPT)
* Voice assistants (like Siri)
* Google Translate (Google Translate)

---

## 🔹 Basic NLP Pipeline

### 1. Text Preprocessing

Cleaning raw text:

* Lowercasing
* Removing punctuation
* Stopwords removal

### 2. Tokenization

Breaking text into words/sentences
👉 “I love NLP” → ["I", "love", "NLP"]

### 3. Stemming / Lemmatization

Reducing words to root form

* Running → Run
* Better → Good (lemmatization)

### 4. Feature Extraction

Convert text → numbers

* Bag of Words
* TF-IDF
* Word Embeddings (Word2Vec, GloVe)

### 5. Model Building

Using ML/DL models:

* Naive Bayes
* Logistic Regression
* RNN, LSTM, Transformers

---

## 🔹 Types of NLP Tasks

* **Sentiment Analysis** → Positive/Negative detection
* **Named Entity Recognition (NER)** → Names, places
* **Machine Translation** → Language conversion
* **Text Summarization** → Short summaries
* **Text Classification** → Spam detection

---

## 🔹 NLP Models Evolution

1. Traditional ML → TF-IDF + ML
2. Deep Learning → RNN, LSTM
3. Transformers → Modern NLP (like BERT, GPT)

---

## 🔹 Real-World Applications

* Chatbots & virtual assistants
* Email spam filtering
* Search engines
* Recommendation systems
* Resume screening

## 🔹 Simple Example (Python)

```python
from nltk.tokenize import word_tokenize

text = "I love NLP"
tokens = word_tokenize(text)

print(tokens)
```
