```python
import nltk
```

```python
corpus = """“Hi, I’m Govardhan Reddy Sidda. I’m currently focusing on Data Science and Generative AI, with a strong foundation in Python and machine learning.I have worked on projects such as RNN-based text classification, which helped me understand deep learning concepts in a practical way. Alongside this, I am continuously improving my problem-solving skills and deep learning knowledge. I am looking for an opportunity where I can apply my skills to real-world problems and grow as a Data Scientist."""
```

```python
corpus
```

```python
### tokenization 
from nltk.tokenize import sent_tokenize # corpus ---> sentence
```

```python
sent_tokenize(corpus)
```

```python
documents = sent_tokenize(corpus)
```

```python
type(documents)
```

```python
for sentence in documents:
    print(sentence)
```

```python
## tokenization 2
# paragraph ---> words
# sentence ---> words

from nltk.tokenize import word_tokenize 
```

```python
word_tokenize(corpus)
```

```python
for sentence in documents:
    print(word_tokenize(sentence))
```

```python
from nltk.tokenize import wordpunct_tokenize
```

```python
wordpunct_tokenize(corpus)
```

```python
from nltk.tokenize import TreebankWordTokenizer
```

```python
tokenizer = TreebankWordTokenizer()
```

```python
tree_tokenize = tokenizer.tokenize(corpus)
tree_tokenize
```

```python
from nltk.tokenize import TreebankWordDetokenizer
```

```python
d_tokenizer = TreebankWordDetokenizer()
```

```python
d_tokenizer.tokenize(tree_tokenize)
```
