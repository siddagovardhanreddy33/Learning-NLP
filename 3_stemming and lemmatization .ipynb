## 🔹 3. Stemming vs Lemmatization

**Goal:** Reduce words to base form

### Stemming (crude):

* Running → run
* Studies → studi ❗

### Lemmatization (accurate):

* Running → run
* Better → good ✅

👉 Lemmatization uses vocabulary + grammar → **more meaningful**


##### Stemming is the process of reducing a word to its root/base form by removing suffixes.

### Simple Definition
- “Stemming is a rule-based process that cuts off prefixes or suffixes from words to obtain a root form, which may not always be a valid English word.”

- stemming is a process of reducing the word to its word stem that affixes to suffixes and prefixes or to the roots of words known as a lemma . stemming in important in nalural language understanding(NLU) and natural language processing (NLP)

### RegexpStemmer class
- NLTK has RegexStemmer class with the help of which we can easily implement Regulaar Expression Stemmer algorithms . it basically takes a single regular expression and removes any prefix or suffix matches the expression 

```python
from nltk.stem import RegexpStemmer
```

```python
# syntax 

#(regexp: Any, min: int = 0) -> RegexpStemmer
#regexp: str or regexp


# A stemmer that uses regular expressions to identify morphological affixes. Any substrings that match the regular expressions will be removed.
```

```python
reg_stemmer = RegexpStemmer('ing$|s$|e$|able$',min=4)
```

```python
reg_stemmer.stem('eating')
```

```python
reg_stemmer.stem('ingeating')
```

```python
for word in words:
    print(word +'---->'+ reg_stemmer.stem(word))
```

# Lemmatization
- lemmatization technique is like stemming.
- the output we will get after lemmatization is called 'lemma' , which is a root word rather than root stem , the output of stemming . after lemmatization we will be getting a valid word that means the same thing  

### WordNet Lemmatizer

- NLTK provides WordNetLemmatizer class which is a wrapper around the wordnet corpus 
- this uses morphy() function to the WordNet corpusReader class to find a lemma

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
lemmatizer
```

```python
lemmatizer.lemmatize()
```

```python
(word: str, pos: str = "n") -> str
word: The input word to lemmatize.


Lemmatize word using WordNet's built-in morphy function. Returns the input word unchanged if it cannot be found in WordNet.

param word : The input word to lemmatize.
type word : str
param pos : The Part Of Speech tag. Valid options are "n" for nouns, "v" for verbs, "a" for adjectives, "r" for adverbs and "s" for satellite adjectives.
param pos : str
return : The lemma of word, for the given pos.
```

```python
# post tag

"""  
Noun -n 
verb - v
adjective - a
adverd - r
"""
```

```python
lemmatizer.lemmatize('going', pos='a')
```

```python
lemmatizer.lemmatize('going', pos='n')
```

```python
lemmatizer.lemmatize('going', pos='v')
```

```python
lemmatizer.lemmatize('going', pos='r')
```

```python
for i in corpus:
    print(i+ '-->' +lemmatizer.lemmatize(i , pos='v') + ' --- verb')
    print(i+ '-->' +lemmatizer.lemmatize(i , pos='a') + ' --- adjective ')    
    print(i+ '-->' +lemmatizer.lemmatize(i , pos='r')+ ' --- adverb')
    print(i+ '-->' +lemmatizer.lemmatize(i , pos='n')+ ' --- noun')
    print('---'*5)
```
